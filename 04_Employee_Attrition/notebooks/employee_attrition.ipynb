#import required libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

#Load the dataset

```python
df = pd.read_csv("../data/HR-Employee-Attrition-All.csv")
```

#Convert Target Variable into Numerical Format

```python
df["Target"] = df["Attrition"].map({
    "Yes": 1,
    "No": 0
})
```

#Check Target Variable Distribution

```python
df["Target"].value_counts()
```

#Drop Original Target Column

```python
df.drop(columns=["Attrition"], inplace=True)
```

#Import ScorecardPy Library for WOE and IV

```python
import scorecardpy as sc
```

#Calculate Weight Of Evidence

```python
bins = sc.woebin(df, y="Target")
```

#Apply WOE Transformation

```python
df_woe = sc.woebin_ply(df, bins)
df_woe.head()
```

#Display Information Values (IV) of Features

```python
iv_df = sc.iv(df, y="Target")
iv_df
```

#Removed features with IV < 0.02, Also removed identifier and constant columns

```python
df_woe = df_woe.drop(columns=[
    "StandardHours",
    "EmployeeCount",
    "Over18",
    "PerformanceRating_woe",
    "Education_woe",
    "EmployeeNumber_woe",
    "Gender_woe",
    "MonthlyRate_woe"
])
```

#Import Library for VIF Calculation

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

#Select Independent variable

```python
X = df_woe.drop(columns="Target")
```

#Calculate VIF For Each Feature

```python
vif = pd.DataFrame()

vif["feature"] = X.columns

vif["VIF"] = [
    variance_inflation_factor(X.values,i)
    for i in range(X.shape[1])
]

vif
```

#Define y

```python
y = df_woe["Target"]
```

#Split Data into Training and Testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size = 0.25,
    random_state=42,
    stratify=y
)
```

#Import Logistic Regression and GridSearchCV

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

#Create Logistic Regression Model

```python
log_model = LogisticRegression(max_iter=1000)
```

#Define Hyperparameter Grid

```python
param_grid = {
    "C" : [0.01, 0.1, 1, 10],
    "solver" : ["liblinear", "lbfgs"]
}
```

#Perform Hyperparameter Tuning using GridSearchCV

```python
grid_search = GridSearchCV(
    estimator=log_model,
    param_grid=param_grid,
    cv=5,
    scoring="f1",
    n_jobs=-1
)
```

#Train Logistic Regression Model

```python
grid_search.fit(X_train, y_train)
```

#Display Best Hyperparameters

```python
print("Best Parameters :", grid_search.best_params_)
print("Best F1 Score :", grid_search.best_score_)
```

#Predict Target Values on Test Data

```python
y_pred = grid_search.predict(X_test)
```

#Import Model Evaluation Metrics

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

#Evaluate Logistic Regression Model

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Accuracy :", accuracy)
print("Precision :", precision)
print("Recall :", recall)
print("F1 :", f1)
```

#Display Confusion Matrix

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

#Display Classification Report

```python
print(classification_report(y_test, y_pred))
```

#Import Decision Tree and GridSearchCV

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

#Create Decision Tree Classifier

```python
dt_model = DecisionTreeClassifier(random_state=42)
```

#Define Hyperparameters for Decision Tree

```python
param_grid = {
    "criterion" : ["gini", "entropy"],
    "max_depth" : [3, 5, 7, 10],
    "min_samples_split" : [2, 5, 10],
    "min_samples_leaf" : [1,2, 4]
}
```

#Perform Hyperparameter Tuning using GridSearchCV

```python
grid_search = GridSearchCV(
    estimator=dt_model,
    param_grid=param_grid,
    cv=5,
    scoring="f1",
    n_jobs=-1
)
```

#Train Decision Tree Model

```python
grid_search.fit(X_train, y_train)
```

#Display Best Hyperparameters

```python
print("Best Parameter :", grid_search.best_params_)
print("Best F1 Score :", grid_search.best_score_)
```

#Predict Target Values on Test Dataset

```python
y_pred = grid_search.predict(X_test)
```

#Import Evaluation Metrics

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix,classification_report
```

#Evaluate Decision Tree Model

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Accuracy :", accuracy)
print("Precision :", precision)
print("Recall :", recall)
print("F1 :", f1)
```

#Display Confusion Matrix

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

#Display Classification Report

```python
print(classification_report(y_test, y_pred))
```

#Display Feature Importance

```python
importance = pd.DataFrame({
    "Feature": X.columns,
    "Importance": grid_search.best_estimator_.feature_importances_
})

importance = importance.sort_values(
    by="Importance",
    ascending=False
)

print(importance.head(10))
```

### This tells us which features contributed the most to the Decision Tree.

Visualize Decision Tree

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(20,10))

plot_tree(
    grid_search.best_estimator_,
    feature_names = X.columns,
    class_names=["no", "yes"],
    filled=True,
    rounded=True,
    fontsize=8
)

plt.show()
```

## Final Conclusion

### Model Comparison

Both Logistic Regression and Decision Tree models were trained and evaluated on the Employee Attrition dataset after performing data preprocessing, roll rate bucket creation, target creation, Weight of Evidence (WOE) encoding, Information Value (IV) analysis, Variance Inflation Factor (VIF) analysis, and hyperparameter tuning using GridSearchCV.

Logistic Regression achieved an Accuracy of **86.68%**, Precision of **63.16%**, Recall of **40.68%**, and an F1 Score of **49.48%**. In comparison, Decision Tree achieved an Accuracy of **82.07%**, Precision of **41.03%**, Recall of **27.12%**, and an F1 Score of **32.65%**.

Since Logistic Regression achieved better performance across all evaluation metrics, particularly Accuracy and F1 Score, it is selected as the final model for this project. These results indicate that Logistic Regression generalized better on the Employee Attrition dataset after feature engineering and model tuning.
